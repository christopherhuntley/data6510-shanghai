## **3. Create a new SQLite database and import the CSV files into import tables.**

- You will likely need to borrow and modify code from previous homeworks. 
- Name the database `HW9_GarmentTags.db`. Where to put it is up to you. 
- Name the tables for the garment tag data `GarmentTags_import` and the service type data `ServiceTypes_import`. 

## **4. Write and debug SQL DDL for the tables in the ERD.** 
- Remember, we're using SQLite, not MySQL. Use the right data types. You may want to review SQLite types and keys again to be sure. 
- Name each table in the plural form, with lowercase letters and underscores.
- Name each column with lowercase letters and underscores. 
- Each PK should be autonumbered *except* `service_id`,  which should use `SMALLINT` data type to [avoid aliasing the `rowid` column](https://www.sqlitetutorial.net/sqlite-primary-key/).  
- You will need to create the tables in the right order; otherwise the FKs constraints won't work.
- Use `DROP TABLE IF EXISTS ...` statements to clear out each table before the `CREATE ...` statements. 


## **5. Populate the service types dimension table.**
The data in the `ServiceTypes_import` table is pre-programmed into the point of sale system. The numbers correspond to the keys on the cash register. 

## **6. Populate remaining tables from `GarmentTags_import`.**
- Remember that the order matters. 
- Where possible, use the numeric dimension identifier columns that already exist in the data rather than generating new keys values. 
- You will need to generate surrogate key values for `brand_id`. Since `brand_id` does not exist in the source data you will need to use an alternate key for the joins when populating the table.
